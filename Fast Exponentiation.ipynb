We can calculate $a ^ {b}$ mod $c$ even for large values of $a$, $b$, $c$ using the fast exponentiation algorithm. Even powers are halved with squaring, and we subtract one from odd powers to make them even. This way we can calculate large powers like $123 ^ {987}$ mod $135$.

```python
def is_even(number):
    return(number % 2 == 0)

def calculate_exponential(base, exponent, modulus):
    if exponent == 0:
        return(1)
    else:
        if is_even(exponent):
            sub_exponential = calculate_exponential(base, exponent // 2, modulus)
            return(sub_exponential * sub_exponential % modulus)
        else:
            sub_exponential = calculate_exponential(base, exponent - 1, modulus)
            return(base * sub_exponential % modulus)

calculate_exponential(123, 987, 135)
```
