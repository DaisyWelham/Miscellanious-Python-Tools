```python
import random as r
r.seed(1)

xData = [r.randint(0, 10) for x in range(10)]
yData = [r.randint(0, 10) for y in range(10)]
print(xData, yData)

#E((X - E(X))(Y - E(Y)))
def covariance(X, Y):
    xMean = sum(X) / len(X)
    yMean = sum(Y) / len(Y)
    return(sum([(X[i] - xMean) * (Y[i] - yMean) for i in range(len(X))]) / len(X))

covariance(xData, xData)
```
